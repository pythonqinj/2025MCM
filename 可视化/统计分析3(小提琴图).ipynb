```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def plot_cluster_violin(df, feature, colors=None):
    """
    输入数据df和特征名，绘制不同类簇的小提琴图，支持动态类簇数量，并将特征名显示为图标题。
    并手动在图中标注中位数和均值。
    
    参数:
    - df: 输入的数据框，必须包含'cluster'列和要绘制的特征列
    - feature: 要绘制的特征名 (字符串)
    - colors: 可选，手动传入的颜色列表，用于控制每个类簇的颜色
    """
    # 检查数据中是否包含'cluster'列和指定特征列
    if 'cluster' not in df.columns or feature not in df.columns:
        raise ValueError("数据框必须包含'cluster'列和指定特征列")
    
    # 获取不同的类簇数量
    clusters = df['cluster'].unique()
    num_clusters = len(clusters)

    # 如果用户未传递颜色列表，则使用默认的tab10颜色映射
    if colors is None:
        colors = sns.color_palette('tab10', num_clusters)  # 使用Seaborn的颜色映射
    elif len(colors) > num_clusters:
        colors = colors[:num_clusters]  # 只使用所需数量的颜色

    # 创建小提琴图
    plt.figure(figsize=(10, 6))  # 设置图表大小
    sns.violinplot(x='cluster', y=feature, hue='cluster', data=df, palette=colors,bw=0.5,inner=None, legend=False,cut = 0)  # 使用seaborn绘制小提琴图，不显示内部元素

    # 计算中位数和均值，并在图中标注
    for i, cluster in enumerate(clusters):
        cluster_data = df[df['cluster'] == cluster][feature]  # 获取每个类簇的特征数据
        mean_val = cluster_data.mean()  # 计算均值
        median_val = cluster_data.median()  # 计算中位数

        # 在图中标注中位数
        plt.scatter(i, median_val, color='white', marker='o', s=20)
        plt.text(i, median_val, f'{median_val:.2f}', fontsize=10, ha='center', va='bottom', color='white')  # 显示中位数值
        # 在图中标注均值
        plt.scatter(i, mean_val, color='black', marker='D', s=20)
        plt.text(i, mean_val, f'{mean_val:.2f}', fontsize=10, ha='center', va='bottom', color='black')  # 显示均值值

    # 设置 X 轴标签
    plt.xticks(ticks=range(num_clusters), labels=[f'Cluster {int(cluster)}' for cluster in clusters])

    # 设置 Y 轴标签
    plt.ylabel(f'{feature} Values')  # 设置y轴标签，显示特征名

    # 设置图表标题为特征名
    plt.title(feature, loc='center', fontsize=20)  # 设置图标题为特征名

    # 添加图例
    handles = [plt.Line2D([0], [0], marker='o', color='w', label=f'Cluster {int(cluster)}', markerfacecolor=colors[i], markersize=10) for i, cluster in enumerate(clusters)]
    plt.legend(handles=handles, loc='upper right')

    # 显示图表
    plt.tight_layout()  # 自动调整子图参数，避免重叠
    plt.show()  # 显示图表

# 加载Excel文件
file_path = r'D:\Desktop\SITP\车辆sitp\code\聚类分析\sitp聚类\统计数据\GMM_for_2_1028.xlsx'  # 文件路径
df = pd.read_excel(file_path, sheet_name='Sheet1')  # 读取Excel文件中的Sheet1表格

# 只选择用于聚类的数值特征（删除 'new_id', 'track_id', 'old_id', 'stationary_time_in_region'）
df = df.drop(columns=['new_id', 'track_id', 'old_id'])

# 自定义颜色
custom_colors = ['red', 'blue', 'green', 'purple', 'orange']  # 自定义的颜色列表

# 遍历所有特征并绘制小提琴图
for feature in df.columns:
    if feature != 'cluster':  # 排除cluster列本身
        plot_cluster_violin(df, feature, colors=custom_colors)
```
